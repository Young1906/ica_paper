```python
X = np.random.randn(10, 1000)
```

```python
plt.rcParams["figure.figsize"] = [10, 5]
def plot_eeg(sig):
    fig = plt.figure()
    n, l = sig.shape # n: number of channels / components; l is lenght of the signal
    t = np.linspace(0,l/500, l)
    
    #Number of axes
    n_ = n + 2 # one for each component, first axes is used for annotation, last for scale
    
    for i in range(n_):
        
        l, b, w, h = 0, 1-(i+1)/n_, 1, 1/n_
        _sig = sig[i,:]
        _min, _max = np.min(_sig), np.max(_sig)
        
        # Signal
        ax = fig.add_axes([l, b, w, h], frameon=False)
        ax.plot(t, _sig, color='black', linewidth=.5)
        
        # Channel names
        ax.text(-.1,.5, "Test",verticalalignment='center')
        
        # Other options
        ax.axis('off')
        
        # Scale
        if i == (n-1):
            l, b, w, h = 1, 1-(i+1)/n, 1 / 15, 1/n
            scale = fig.add_axes([l, b, w, h])
            scale.axvline(0, color="black")
#             scale.axhline(0, color="black")
            scale.text(0.1, .5, f"{np.around(_max - _min,2)}", verticalalignment='center')
            scale.axis("off")
    return fig
```

```python
plot_eeg(X)
plt.yscale("linear")
plt.show()
```
